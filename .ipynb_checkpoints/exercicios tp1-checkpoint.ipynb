# **1.2.** Cálculo Recursivo

```python
def recursive(n):
    if n == 1:
        return 1
    else:
        return (4*n-2)/(n+1)*recursive(n-1)
    
print(recursive(100))
```

Tempo de execução: 0.0s

```python

def factorial(n):
    if n == 1.:
        return 1
    else:
        return n*factorial(n-1)
    
print(factorial(200))
```

```python
def fixedFactorial(n):
    if isinstance(n, float):
        n = int(n)
    if n == 1 or n ==0:
        return 1
    else:
        return n*factorial(n-1)

fixedFactorial(20)
```

# **1.4.** Constante de Mandelung

### Usando while loop

```python
N = 100
f = 0

#contribuição dos eixos
i,j,k = 0,0,1
while k < N:
    f += 6*(-1)**k/abs(k)
    k +=1

#contribuição dos planos
i,j,k = 0,1,1
while j < N:
    while k < N:
        f += 12*(-1)**(i+j+k)/(math.sqrt(i**2+j**2+k**2))
        k += 1
    j += 1
    k = 1
    
#contribuição do miolo
i,j,k = 1,1,1
while i < N:
    while j < N:
        while k < N:
            f += 8*(-1)**(i+j+k)/(math.sqrt(i**2+j**2+k**2))
            k += 1
        j += 1
        k = 1
    i += 1
    j = 1
    k = 1



print(f)
```

Tempo de execução: 0.8s

### Usando for loop - ❗**MAIS EFICIENTE**

```python


def mandelung(L):
    M = 0

    #contribuição dos eixos
    for i in range(1,L+1):
        M += 6*(-1)**(i)/abs(i)
    
    #contribuição dos planos
    for i in range(1,L+1):
        for j in range(1,L+1):
            M += 12*(-1)**(i+j)/math.sqrt(i**2+j**2)
    
    #contribuição do miolo
    for i in range(1,L+1):
        for j in range(1,L+1):
            for k in range(1,L+1):
                M += 8*(-1)**(i+j+k)/math.sqrt(i**2+j**2+k**2)

    return M

print(mandelung(N))
    
```

Tempo de execução: 0.4s

# **1.5.** Coeficientes Binomiais

### 1

```python
def binomial(n,k):
    nf = fixedFactorial(n)
    kf = fixedFactorial(k)
    nkf = fixedFactorial(n-k)
    return int(nf/(kf*nkf))

print(binomial(10,9))
```

Tempo de execução: 0.0s

### 2

```python
N = 20
n = 0
while n <= N:
    k = 0
    row = ""
    while k <= n:
        row += f"{binomial(n,k)} "
        k +=1
    print(row)
    n +=1
    
```

Tempo de execução: 0.0s

# **1.7.**  Erros de truncagem - Equação quadrática

```python
#1

a = .001
b = 1000
c = .001


def quadratic(a,b,c):
    if b**2-4*a*c < 0: 
        return "Sol. Complexa"
    x1 = (-b - math.sqrt(b**2-4*a*c))/(2*a)
    x2 = (-b + math.sqrt(b**2-4*a*c))/(2*a)

    return [x1,x2]

print(quadratic(a,b,c))
```

Tempo de execução: 0.0s

```python
#2
def quadratic2(a,b,c):
    if b**2-4*a*c < 0: 
        return "Sol. Complexa"
    x1 = 2*c/(-b+math.sqrt(b**2-4*a*c))
    x2 = 2*c/(-b-math.sqrt(b**2-4*a*c))
    return [x1,x2]

print(quadratic2(a,b,c))
```

Tempo de execução: 0.0s

O que acontece é que, em módulo $b$ é muito maior do que $\sqrt{b^2+4ac}$ e ao somar os dois existe um erro muito grande nos ultimos dígitos. Como as ordens de grandeza destes números são tão diferentes, o computador perde precisão em algarismos mais baixos.

```python

print("Método 1")
print(-b - math.sqrt(b**2-4*a*c))
print(-b + math.sqrt(b**2-4*a*c))

print("Método 2")
print(-b+math.sqrt(b**2-4*a*c))
print(-b-math.sqrt(b**2-4*a*c))
```

Tempo de execução: 0.0s

Com este teste nota-se que o método 1 é mais preciso a calcular a raiz positiva e o método 2 é melhor para calcular a raiz negativa.

```python
def quadratic_safe(a,b,c):
    if b**2-4*a*c < 0: 
        return "Sol. Complexa"
    x1 = 2*c/(-b+math.sqrt(b**2-4*a*c))
    x2 = (-b + math.sqrt(b**2-4*a*c))/(2*a)

    return [x1,x2]

print(quadratic_safe(a,b,c))
```

Tempo de execução: 0.0s

# **1.8.**  Erros de truncagem - Derivada Simplista

### 1

```python
def f(x, delta):
    f = x*(x-1)
    derivative = ((x+delta)*(x+delta-1)-f)/delta
    return [f, derivative]

delta = 1e-2
print(f(1, delta))
```

Tempo de execução: 0.0s

Analíticamente $\frac{df(x)}{dx} = (x-1) + x = 2x-1$.

Para $x=1$, $\left.\frac{d f(x)}{dx}\right|_{x=1} = 1$.

Os valores não são iguais, pois o valor que foi calculado é o decilve da reta que liga os pontos $x = 1$ e $x = 1.01$. Não é o declive da reta tangente, mas sim uma aproximação.

### 2

```python
delta = 1e-4
print("δ=10e-4:", f(1,delta))

delta = 1e-6
print("δ=10e-6:", f(1,delta))

delta = 1e-8
print("δ=10e-8:", f(1,delta))

delta = 1e-10
print("δ=10e-10:", f(1,delta))

delta = 1e-12
print("δ=10e-12:", f(1,delta))

delta = 1e-14
print("δ=10e-14:", f(1,delta))
```

Tempo de execução: 0.0s

O valor é cada vez mais preciso até $\delta =10^{-8}$, mas a partir daí começam a existir grandes erros de truncagem.
